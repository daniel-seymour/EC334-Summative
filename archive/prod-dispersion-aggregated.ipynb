```python
import pandas as pd
import numpy as np
```

```python
filepath = '/Users/danielseymour/Developer/EC334-Summative/raw_data/productivity_data/total_factor_productivity.csv'

df = pd.read_csv(filepath)
df.head()
```

```python
# Convert object columns to numeric
columns_to_convert = ['d9990', 'd1001', 'd9990*', 'd1001*']

for col in columns_to_convert:
   df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.describe()
```

```python
# Select only the columns with asterisks
asterisk_cols = ['sd*', 'd7525*', 'd9010*', 'd9990*', 'd1001*']

# Get descriptive statistics for these columns
desc_stats = df[asterisk_cols].describe()

# Transpose to have variables as rows (standard format for econ papers)
desc_stats = desc_stats.T

# Select key statistics typically shown in economics papers
desc_stats = desc_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]

# Rename columns to be more formal
desc_stats.columns = ['N', 'Mean', 'Std. Dev.', 'Min', 'p25', 'Median', 'p75', 'Max']

# Create nicer row labels
row_labels = {
    'sd*': 'Productivity Dispersion (SD)',
    'd7525*': 'Productivity Dispersion (75-25)',
    'd9010*': 'Productivity Dispersion (90-10)', 
    'd9990*': 'Productivity Dispersion (99-90)',
    'd1001*': 'Productivity Dispersion (10-1)'
}
desc_stats.index = desc_stats.index.map(row_labels)

# Round to 3 decimal places
desc_stats = desc_stats.round(3)

# Convert to LaTeX
latex_table = desc_stats.to_latex(
    float_format="%.3f",
    caption="Descriptive Statistics: Employment-Weighted Productivity Dispersion Measures (4-digit NAICS)",
    label="tab:prod_dispersion_stats",
    column_format='lrrrrrrrr'  # left-align first column, right-align numbers
)

# Print the LaTeX code
print(latex_table)

# Save to file - CORRECTED PATH WITH FILENAME
with open('/Users/danielseymour/Developer/EC334-Summative/tables/prod_dispersion_stats.tex', 'w') as f:
    f.write(latex_table)

# Also display the dataframe
print("\nDescriptive Statistics:")
print(desc_stats)
print("\nLaTeX table saved to: /Users/danielseymour/Developer/EC334-Summative/tables/prod_dispersion_stats.tex")
```

```python


# Create 3-digit NAICS from 4-digit
df['naics3'] = df['naics4'].astype(str).str[:3]

# Select all employment-weighted (asterisk) variables
weighted_vars = ['sd*', 'd7525*', 'd9010*', 'd9990*', 'd1001*']

# Aggregate to 3-digit level using simple average of weighted measures
df_3digit = df.groupby(['year', 'naics3'])[weighted_vars].mean().reset_index()

# Rename columns for clarity (remove asterisk and add _weighted suffix)
new_column_names = ['year', 'naics3'] + [col.replace('*', '_weighted') for col in weighted_vars]
df_3digit.columns = new_column_names

# Sort by year and naics3
df_3digit = df_3digit.sort_values(['year', 'naics3']).reset_index(drop=True)

# Display summary
print("New dataframe shape:", df_3digit.shape)
print("\nColumns in new dataframe:")
print(df_3digit.columns.tolist())

print("\nFirst 10 rows:")
print(df_3digit.head(10))

# Save the aggregated data
output_path = '/Users/danielseymour/Developer/EC334-Summative/processed_data/prod_dispersion_3digit.csv'
df_3digit.to_csv(output_path, index=False)
print(f"\nData saved to: {output_path}")
```

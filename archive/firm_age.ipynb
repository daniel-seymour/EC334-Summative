```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/danielseymour/Developer/EC334-Summative/raw_data/dynamism_data/bds2022_vcn3_fa.csv')

# Display basic information and statistics about the dataset
print(df.info())
print(df.describe(include='all'))
print(df.head())
```

```python
# 1. Convert numeric columns from strings to proper numeric types
# First, replace 'X' values with NaN (these represent suppressed data)
df = df.replace('X', pd.NA)

# 2. Convert columns to numeric types
numeric_columns = ['firms', 'estabs', 'emp', 'estabs_entry', 'estabs_exit', 
                  'job_creation', 'job_destruction', 'net_job_creation',
                  'firmdeath_firms', 'firmdeath_estabs', 'firmdeath_emp']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 3. Filter for age 0 firms (new firms) and select relevant columns
new_firms_df = df[df['fage'] == 'a) 0'][['year', 'firms', 'estabs', 'emp']]

# 4. Set year as the index for easy reference
new_firms_df = new_firms_df.set_index('year')

# 5. Rename columns to be more descriptive
new_firms_df = new_firms_df.rename(columns={
    'firms': 'new_firms',
    'estabs': 'new_establishments',
    'emp': 'new_firm_employment'
})

# 6. Sort by year
new_firms_df = new_firms_df.sort_index()

# Then calculate total firms by year
total_firms_by_year = df.groupby('year')['firms'].sum()

# Create the new firms dataframe
new_firms_df = df[df['fage'] == 'a) 0'][['year', 'firms']].set_index('year')
new_firms_df = new_firms_df.rename(columns={'firms': 'new_firms'})

# Add total firms and calculate the new firm rate
new_firms_df['total_firms'] = total_firms_by_year
new_firms_df['new_firm_rate'] = new_firms_df['new_firms'] / new_firms_df['total_firms'] * 100

# Display the result
print(new_firms_df)
```

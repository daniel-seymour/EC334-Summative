```python
# Start with simple approach of two-digit NAICS codes, dropping any rows that don't have two digits and removing the change from previous year

import pandas as pd
import numpy as np

# Step 1: Filter the TFP dataset to keep only index values (not percentage changes)
tfp_filtered_df = tfp_long_df[tfp_long_df['Units'] == 'Index (2017=100)']

# Step 2: Filter to keep only standard 2-digit NAICS codes
# Function to check if a NAICS code is a standard 2-digit code
def is_standard_2digit(naics):
    naics_str = str(naics).strip()
    
    # Check if it's exactly 2 digits
    return naics_str.isdigit() and len(naics_str) == 2

# Apply filter to keep only standard 2-digit NAICS codes
tfp_filtered_df = tfp_filtered_df[tfp_filtered_df['NAICS'].apply(is_standard_2digit)]

# Step 3: Convert TFP values to numeric
tfp_filtered_df['tfp_value'] = pd.to_numeric(tfp_filtered_df['tfp_value'], errors='coerce')

# Step 4: Extract 2-digit NAICS from BDS data
bds_df['naics2'] = bds_df['vcnaics4'].astype(str).str[:2]

# Step 5: Convert all relevant columns in BDS data to numeric
# First identify columns that should be numeric
numeric_cols = [col for col in bds_df.columns if col not in ['year', 'naics2', 'vcnaics4']]

# Convert to numeric
for col in numeric_cols:
    bds_df[col] = pd.to_numeric(bds_df[col], errors='coerce')

# Step 6: Aggregate BDS data to 2-digit level
# Calculate weights for weighted averages
bds_df['weight'] = bds_df['emp'].fillna(0)

# Define column types
count_cols = [col for col in numeric_cols if 'rate' not in col and col != 'weight']
rate_cols = [col for col in numeric_cols if 'rate' in col]

# Basic aggregation for count columns
agg_dict = {col: 'sum' for col in count_cols}
grouped = bds_df.groupby(['year', 'naics2'])
bds_agg = grouped.agg(agg_dict)

# Weighted averages for rate columns
for col in rate_cols:
    # Skip if column doesn't exist or has all missing values
    if col not in bds_df.columns or bds_df[col].isna().all():
        continue
        
    # Calculate weighted values
    bds_df[f'{col}_weighted'] = bds_df[col] * bds_df['weight']
    
    # Sum weighted values and weights
    weighted_sums = grouped[f'{col}_weighted'].sum()
    weights = grouped['weight'].sum()
    
    # Calculate weighted average with epsilon to avoid division by zero
    epsilon = 1e-10
    bds_agg[col] = weighted_sums / (weights + epsilon)

# Reset index
bds_agg = bds_agg.reset_index()

# Step 7: Merge the datasets
merged_df = pd.merge(
    bds_agg,
    tfp_filtered_df,
    how='inner',
    left_on=['year', 'naics2'],
    right_on=['year', 'NAICS']
)

# Step 8: Check the result
print(f"Merged dataset shape: {merged_df.shape}")
print(f"Number of industries (NAICS codes): {merged_df['naics2'].nunique()}")
print(f"NAICS codes in merged data: {sorted(merged_df['naics2'].unique())}")
print(f"Year range: {merged_df['year'].min()} to {merged_df['year'].max()}")

# Display the first few rows
print("\nMerged dataset preview:")
print(merged_df.head())

# Optional: Save to CSV
# merged_df.to_csv('merged_bds_tfp_standard_2digit.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Step 1: Extract 2-digit NAICS codes from the 4-digit codes in the BDS dataset
bds_df['naics2'] = bds_df['vcnaics4'].astype(str).str[:2]

# Step 2: Convert all numeric columns in bds_df to proper numeric type
# First identify all rate columns
rate_cols = [col for col in bds_df.columns if 'rate' in col]
count_cols = [col for col in bds_df.columns if col not in rate_cols + ['year', 'naics2', 'vcnaics4']]

# Convert all numeric columns to proper numeric type
for col in rate_cols + count_cols:
    bds_df[col] = pd.to_numeric(bds_df[col], errors='coerce')

# Step 3: Calculate weights for weighted averages (using employment as weight)
bds_df['weight'] = bds_df['emp'].fillna(0)  # Replace any NaN with 0

# Step 4: Reshape the TFP dataset from wide to long format
tfp_long_df = pd.melt(
    sectoral_tfp_df,
    id_vars=['NAICS', 'Industry', 'Basis', 'Measure', 'Units'],
    value_vars=[str(year) for year in range(1987, 2024)],
    var_name='year',
    value_name='tfp_value'
)

# Convert year column to numeric for proper joining
tfp_long_df['year'] = pd.to_numeric(tfp_long_df['year'])

# Convert tfp_value to numeric
tfp_long_df['tfp_value'] = pd.to_numeric(tfp_long_df['tfp_value'], errors='coerce')

# Clean up and standardize the NAICS codes in the TFP dataset
tfp_long_df['naics2'] = tfp_long_df['NAICS'].astype(str).str.extract(r'^(\d{2})').iloc[:, 0]

# Step 5: Aggregate the BDS data to the 2-digit NAICS level
# Define aggregation methods for different column types
agg_dict = {}

# For count columns, we sum
for col in count_cols:
    agg_dict[col] = 'sum'

# For rate columns, we'll compute weighted average later
# (Skip them in the initial aggregation)

# Perform the basic aggregation for count columns
grouped = bds_df.groupby(['year', 'naics2'])
bds_agg = grouped.agg(agg_dict)

# Step 6: Calculate weighted averages for rate columns manually
for col in rate_cols:
    # Calculate the weighted values (skip NaN values)
    bds_df[f'{col}_weighted'] = bds_df[col] * bds_df['weight']
    
    # Sum the weighted values and the weights
    weighted_sums = grouped[f'{col}_weighted'].sum()
    weights = grouped['weight'].sum()
    
    # Calculate the weighted average and handle division by zero
    # Add a small epsilon to weights to avoid division by zero
    epsilon = 1e-10
    bds_agg[col] = weighted_sums / (weights + epsilon)

# Reset index to get regular columns
bds_agg = bds_agg.reset_index()

# Step 7: Merge the aggregated BDS data with the TFP data
merged_df = pd.merge(
    bds_agg,
    tfp_long_df,
    how='inner',
    left_on=['year', 'naics2'],
    right_on=['year', 'naics2']
)

# Step 8: Check the merged dataset
print(f"Merged dataset shape: {merged_df.shape}")
print(f"Aggregated BDS dataset shape: {bds_agg.shape}")
print(f"Transformed TFP dataset shape: {tfp_long_df.shape}")

# View the first few rows of the merged dataset
print("\nMerged dataset preview:")
print(merged_df.head())

# Optional: Filter for specific TFP measures if needed
# If you want to keep only Total factor productivity measure:
merged_tfp_df = merged_df[merged_df['Measure'] == 'Total factor productivity']

# Save the merged dataset to a new CSV file if desired
# merged_df.to_csv('merged_bds_tfp_data.csv', index=False)
```
